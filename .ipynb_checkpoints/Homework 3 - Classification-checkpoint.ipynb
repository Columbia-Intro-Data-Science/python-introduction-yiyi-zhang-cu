```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import numpy as np
from sklearn.metrics import roc_curve, auc
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from IPython.display import Image
from sklearn.externals.six import StringIO
import pydotplus
from sklearn.tree import DecisionTreeRegressor
from collections import Counter
```

1) Load in the dataset `https://d1pqsl2386xqi9.cloudfront.net/notebooks/Default.csv` into a pandas dataframe

```python
#import pandas as pd
df = pd.read_csv('https://d1pqsl2386xqi9.cloudfront.net/notebooks/Default.csv')
df = df.drop(columns='Unnamed: 0')
df.head()
```

2) Convert `default` and `student` to categorical variables.

```python
df['default'] = df['default'].map({'Yes': 1, 'No': 0})
df['student'] = df['student'].map({'Yes': 1, 'No': 0})
df.head()
```

3) Make boxplots of all of the variables and a scatter plot of of (balance,income) with the color
of the outcome of `default` labeled red if yes and blue if no. 

```python
#import matplotlib.pyplot as plt
#import seaborn as sns
plt.figure(figsize=(15,5))
plt.subplot(1, 3, 1)
df.boxplot()
plt.subplot(1, 3, 2)
sns.boxplot(x="student", y="balance", hue="default", data=df)
plt.subplot(1, 3, 3)
sns.boxplot(x="student", y="income", hue="default", data=df)
plt.show()
```

```python
d0 = (df[df.default==0].balance, df[df.default==0].income)
d1 = (df[df.default==1].balance, df[df.default==1].income)
data = (d0, d1)
colors = ("blue", "red")
groups = ("No_Default", "Yes_Default") 

fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(1, 1, 1)
for data, color, group in zip(data, colors, groups):
    x, y = data
    ax.scatter(x, y, alpha=0.8, s=8, c=color, label=group)

plt.title("Income vs. Balance")
plt.xlabel("Balance")
plt.ylabel("Income")
plt.legend(loc=1)
plt.show()
```

4) Train a model to predict if a student defaults on their loan or not using LogisticRegression based soley
on the `balance` and `income` variables. Plot the decision boundary in the scatter plot you made in 3) and compute the AUC ROC.

```python
#from sklearn.model_selection import train_test_split
#from sklearn.linear_model import LogisticRegression
#import numpy as np

# split train/test set
df2 = df[df['student'] == 1]
X = df2[['balance','income']]
y = df2['default']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state=123)

# train model 
Logit = LogisticRegression(penalty='l1')
fit = Logit.fit(X_train, y_train)

# get model parameters
a = fit.coef_[0]
b = fit.intercept_

fig, [ax1, ax2] = plt.subplots(nrows=1, ncols=2, figsize=(10,5))

# plot scatter plot
ax1.scatter(df["balance"], df["income"], s=8, c = df["default"].apply(lambda x:colors[x]))

# plot decision boundary
umin, umax = ax1.get_xlim()
vmin, vmax = ax1.get_ylim()
u = np.arange(umin, umax)
v = (b+a[0]*u)/(-a[1])
ax1.plot(u,v)
ax1.set_ylim(vmin,vmax)
ax1.set_title("Income vs. Balance with Decision Boundary")
ax1.set_xlabel("Balance")
ax1.set_ylabel("Income")

# compute AUC ROC
#from sklearn.metrics import roc_curve, auc
y_score = fit.decision_function(X_test)
fpr, tpr, thre = roc_curve(y_test,y_score)
auc_roc = auc(fpr, tpr)

ax2.plot(fpr, tpr, label='ROC curve (AUC = %0.2f)'%auc_roc)
ax2.plot([0,1], [0,1], linestyle='--')
ax2.set_xlabel('False Positive Rate')
ax2.set_ylabel('True Positive Rate')
ax2.set_title('ROC Curve')
ax2.legend(loc = 'lower right')
plt.show()
```

AUC ROC is 0.95.

5) Repeat 4) but with the student variable included. Optimize using Ridge regression and plot the ranking
of coefficients.  How does performance improve with the added variable?

```python
# split train/test set
X = df[['student','balance','income']]
y = df['default']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state=123)

# train model 
Logit = LogisticRegression(penalty='l1')
fit = Logit.fit(X_train, y_train)

# get model parameters
a = fit.coef_[0]
b = fit.intercept_

fig, [ax1, ax2] = plt.subplots(nrows=1, ncols=2, figsize=(10,5))

# plot scatter plot
ax1.scatter(df["balance"], df["income"], s=8, c = df["default"].apply(lambda x:colors[x]),label="")

# plot decision boundary
umin, umax = ax1.get_xlim()
vmin, vmax = ax1.get_ylim()
u = np.arange(umin, umax)
v = (b+a[1]*u)/(-a[2])
w = (b+a[0]+a[1]*u)/(-a[2])
ax1.plot(u,v,label="No_Student")
ax1.plot(u,w,label="Yes_Student")
ax1.set_ylim(vmin,vmax)
ax1.set_title("Income vs. Balance with Decision Boundary")
ax1.set_xlabel("Balance")
ax1.set_ylabel("Income")
ax1.legend(loc=1)

# compute AUC ROC
from sklearn.metrics import roc_curve, auc
y_score = fit.decision_function(X_test)
fpr, tpr, thre = roc_curve(y_test,y_score)
auc_roc = auc(fpr, tpr)

ax2.plot(fpr, tpr, label='ROC curve (AUC = %0.2f)'%auc_roc)
ax2.plot([0,1], [0,1], linestyle='--')
ax2.set_xlabel('False Positive Rate')
ax2.set_ylabel('True Positive Rate')
ax2.set_title('ROC Curve')
ax2.legend(loc = 'lower right')
plt.show() 
```

```python
#from sklearn.linear_model import Ridge
#from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state=123)

alphas = np.logspace(-5, -1, 100)
score=[]
for alpha in alphas:
    clf = Ridge(alpha = alpha)
    clf.fit(X_train, y_train)
    score.append(clf.score(X_test, y_test))
index = np.array(score).argmax()

clf = Ridge(alpha = alphas[index])
clf.fit(X_train, y_train)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.bar(np.arange(3), clf.coef_, 0.5)
ax.set_xticks(np.arange(3))
ax.set_xticklabels(('student', 'balance','income'))
rects = ax.patches
plt.show()
print ('The coefficients:',clf.coef_)
```

AUC ROC improved by 0.01 with the added variable.

6) Using cross validation, check that the model you built in 5) is robust (generalizes to other random folds of the dataset).
